# 原假设和备择假设的建立

统计推断：如何利用**假设检验**来确定是否应拒绝关于总体参数数值。

在假设检验中，我们首先对总体参数做一个尝试性的假设，该尝试性的假设称为原假设，记作 $H_0$。

将受到挑战的假说作为原假设，零假设的任务就是为了被**推翻**。

然后，定义另一个与原假设内容完全对立的假设，记作 $H_{\alpha}$，称之为备择假设。

将研究中的假设作、目标立为备择假设，如考察新方法是否更好

假设检验的过程，就是根据样本数据，对这两个对立的假设：$H_0$、$H_{\alpha}$ 进行检验的过程。

## 原假设和备择假设形式的小结

假设检验主要有三种形式：前两种为单侧检验 第三种为双侧检验

下侧检验
- 原假设 $H_0: \mu \ge \mu_0$
- 备择假设 $H_{\alpha}: \mu < \mu_0$

上侧检验
- 原假设 $H_0: \mu \le \mu_0$
- 备择假设 $H_{\alpha}: \mu > \mu_0$ 

双侧检验
- 原假设 $H_0: \mu = \mu_0$
- 备择假设 $H_{\alpha}: \mu \neq \mu_0$ 

# 第一类错误和第二类错误 

- 第一类错误：$H_0$ 为真时，拒绝 $H_0$
- 第二类错误：$H_{\alpha}$ 为真时，没有拒绝 $H_0$

|    分类     | 零假设为真 | 备择假设为真 |
| :--------: | :-------: | :-------: |
|  拒绝零假设  | 第一类错误 |  正确决定  |
| 不能拒绝零假设 |  正确决定  | 第二类错误 |

当原假设为真，并且以**等式**形式出现时，犯第一类错误的概率称为检验的显著性水平 $\alpha$ 

在实践中，由负责假设检验的人设定显著性水平。通过选择 $\alpha$ 控制犯第一类错误的概率。

- 如果犯第一类错误的成本很高，则选择小的 $\alpha$ 值。
- 如果犯第一类错误的成本不高，则通常选择较大的 $\alpha$ 值。

在应用中，一般将**只控制第一类错误**的假设检验称为显著性检验，许多假设检验的应用都属于这一类型。

在假设检验的大多数应用中，虽然对发生第一类错误的概率进行了控制，但通常并没有控制发生第二类错误的概率。

因此，如果决定接受 $H_0$，我们**并不能确定**该决策有多大的可信度。

由于显著性检验中**第二类错误的发生具有不确定性**，所以通常统计学家建议我们在叙述中采用“不能拒绝 $H_0$”，而不采用“接受 $H_0$”这种说法。

采用“不能拒绝 $H_0$”这种说法意味着我们对**判断或行动持保留意见**。

实际上，不直接接受 $H_0$，使统计学家**规避了发生第二类错误的风险**。

不论何时，只要未对发生第二类错误的概率加以确定或控制，我们就不能得出接受 $H_0$ 的结论。

这种情形下，只能得出两种可能的结论：
- 不能拒绝 $H_0$
- 拒绝 $H_0$

通过假设检验，我们只能判定一件事：**能否拒绝 $H_0$**，至于接受 $H_0$ 或者 $H_{\alpha}$，不能简单通过假设检验来验证。

虽然在假设检验中通常没有控制第二类错误，但是第二类错误确实是可以控制的。

后面我们将介绍确定和控制第二类错误发生概率的方法。如果能够对第二类错误进行适当的控制，那么基于“接受 $H_0$”的结论所采取的措施就是恰当的。

#  总体均值的检验：$\sigma$ 已知情形

检验统计量：用于确定是否拒绝原假设统计量

总体均值假设检验的检验统计量：$\sigma$ 已知的情形，计算 $z$ 值的公式为：

$$z = \frac {\bar{x}- \mu_0} {\frac{\sigma}{\sqrt{n}}}$$

- $\bar{x}$ 为样本均值
- $\mu_0$ 为原假设均值
- $\sigma$ 为总体标准差
- $n$ 样本数
- $z$ 为检验统计量

## 下册检验
已知
- $\mu_0 = 3$
- $\sigma = 0.18$
- $n = 36$
- $\bar{x} = 2.92$
- 原假设 $H_0: \bar{x} \ge \mu_0$
- 备择假设 $H_{\alpha}: \bar{x} < \mu_0$

则样本标准差

$$\displaystyle \sigma_\bar{x} = \frac{\sigma}{\sqrt{n}} = \frac{0.18}{\sqrt{36}} = 0.03$$

$$z = \frac {\bar{x}- \mu_0} {\sigma_\bar{x}} = \frac {\bar{x}- 3} {0.03}$$

已知 $\bar{x} = 2.92$，计算得 $z = -2.67$，查阅 $z$ 值表得此时的 $p$ 值为 0.0038

$p-$ 值：$p-$ 值是一个概率值，度量样本所提供的证据对原假设的支持程度。$p-$ 值越小说明反对原假设的证据越多

$p-$ 值法的拒绝法则：如果 $p \le \alpha$ 则拒绝 $H_0$

假设显著性水平 $\alpha$ 设置为 0.01，则此时 $p < \alpha$，因此我们拒绝 $H_0$

## 下侧检验的拒绝法则：临界值法

临界值法：临界值法要求我们首先确定被称为临界值的检验统计量的值

临界值是使得我们拒绝原假设的检验统计量的最大值，是对应与实际显著性水平 $\alpha$ 对应的 $z$ 值。

如果$z \ge -z_{\alpha}$， 则拒绝 $H_0$

$-z_{\alpha}$ 为临界值，即标准正态概率分布下侧的面积为 $\alpha$ 时对应的 $z$ 值。

同理，可以推导上侧检验和双侧检验。

临界的显著性水平 $\alpha = 0.01$，如果要适用临界法则，假设 $p = \alpha = 0.01$，查阅对应的 $z = -2.33$。

$$z = \frac {\bar{x}- \mu_0} {\sigma_\bar{x}} = \frac {\bar{x}- 3} {0.03} = -2.33$$

得 $\bar{x} = 2.9301$，由于我们计算得出临界的 $\bar{x}$ 值，因此我们可以快速判断，只要 $\bar{x} < 2.93$ 时，我们都可以拒绝 $H_0$

```python
alpha = 0.01

mu_0 = 3
x_bar = 2.92
sigma = 0.18
n = 36

stdev = sigma/n**0.5

# Our calculated z-score in this example:
p_crit = norm.ppf(q = alpha)
p_val = norm.cdf(x = x_bar, loc = mu_0, scale = stdev)
p_stat = norm.ppf(p_val)
#print(p_crit, p_val, p_stat)
```

```python
def left_tail_test(p_crit, p_stat, p_val):
    x = np.linspace(start = norm.ppf(0.001), stop = norm.ppf(0.999), num = int(1e5))
    hx = norm.pdf(x)
    # Plot the density
    fig, ax = plt.subplots(num = 1, figsize = (10, 8))
    _ = plt.plot(x, hx, color = "black")
    _ = plt.margins(y = 0)
    _ = plt.ylim((0, 0.41))
    # Shade the probability alpha
    _ = plt.fill_between(x[x <= p_crit], hx[x <= p_crit], 
                     edgecolor = "red", linestyle = "-", linewidth = 2,
                     facecolor = clrs.to_rgba(c = (1, 0, 0, 0.2)), zorder = 10)

    # plot the critical value p_c:
    _ = plt.axvline(x = p_crit, color = "red", linestyle = "--")

    # Shade the probability p-value
    _ = plt.fill_between(x[x <= p_stat], hx[x <= p_stat], edgecolor = None,
                     color = clrs.to_rgba(c = (0, 1, 0, 0.4)))

    # plot the observed p-statictic:
    _ = plt.axvline(x = p_stat, color = "darkgreen", linestyle = "--")
    #
    xt = np.append(ax.get_xticks(), [p_crit, p_stat])
    xtl = xt.tolist()
    xtl[-2:] = ["$pc$", "$p_{stat}$"]
    _ = ax.set_xticks(xt)
    _ = ax.set_xticklabels(xtl)
    ax.get_xticklabels()[-2].set_color("red")
    ax.get_xticklabels()[-1].set_color("darkgreen")
    _ = plt.title("p-value for the left-tail test; $H_\\alpha: \\beta_1 < 0$")
    plt.show()
```

```python
left_tail_test(p_crit, p_stat, p_val)
```

## 右侧检验

```python
alpha = 0.05

mu_0 = 3
x_bar = 3.04
sigma = 0.18
n = 36

stdev = sigma/n**0.5

# Our calculated z-score in this example:
p_crit = norm.ppf(q = 1 - alpha)
p_val = norm.cdf(x = x_bar, loc = mu_0, scale = stdev)
p_stat = norm.ppf(p_val)
print(p_crit, p_val, p_stat)
```

```python
def right_tail_test(p_crit, p_stat, p_val):
    x = np.linspace(start = norm.ppf(0.001), stop = norm.ppf(0.999), num = int(1e5))
    hx = norm.pdf(x)
    # Plot the density
    fig, ax = plt.subplots(num = 0, figsize = (10, 8))
    _ = plt.plot(x, hx, color = "black")
    _ = plt.margins(y = 0)
    _ = plt.ylim((0, 0.41))

    # Shade the probability alpha
    _ = plt.fill_between(x[x >= p_crit], hx[x >= p_crit], 
                     edgecolor = "red", linestyle = "-", linewidth = 2,
                     facecolor = clrs.to_rgba(c = (1, 0, 0, 0.2)), zorder = 10)

    # plot the critical value p_c:
    _ = plt.axvline(x = p_crit, color = "red", linestyle = "--")
    
    # Shade the probability p-value
    _ = plt.fill_between(x[x >= p_stat], hx[x >= p_stat], edgecolor = None,
                     color = clrs.to_rgba(c = (0, 1, 0, 0.4)))
    
    # plot the observed t-statictic:
    _ = plt.axvline(x = p_stat, color = "darkgreen", linestyle = "--")
    
    xt = np.append(ax.get_xticks(), [p_crit, p_stat])
    xtl = xt.tolist()
    xtl[-2:] = ["$p_c$", "$p_{stat}$"]
    _ = ax.set_xticks(xt)
    _ = ax.set_xticklabels(xtl)
    ax.get_xticklabels()[-2].set_color("red")
    ax.get_xticklabels()[-1].set_color("darkgreen")
    _ = plt.title("p-value for the right-tail test; $H_{\\alpha}: \\beta_1 > 0$")
    plt.show()
```

```python
right_tail_test(p_crit, p_stat, p_val)
```

## 双侧检验

已知
- $\mu_0 = 295$
- $\sigma = 12$
- $n = 50$
- 原假设 $H_0: \mu = \mu_0$
- 备择假设 $H_{\alpha}: \mu \neq \mu_0$

则样本标准差

$$\displaystyle \sigma_\bar{x} = \frac{\sigma}{\sqrt{n}} = \frac{12}{\sqrt{50}} = 1.7$$

$$z = \frac {\bar{x}- \mu_0} {\sigma_\bar{x}} = \frac {\bar{x}- 295}{1.7}$$

假设我们现在已知 $\bar{x} = 297.6$，计算得 $z = 1.53$，查阅 $z$ 值表得此时的 $p$ 值为 0.937

$P(z \ge 1.53) = 1 - 0.937 = 0.063$

我们将显著性水平 $\alpha$ 设置为 0.05，由于此时是双侧检验，所以我们不能用 $P(z \ge 1.53)$ 和 $\alpha$ 直接对比，还需要计算另一侧的值：

$P(z \le -1.53) = 1 - 0.937 = 0.063$

由于 $P(z \ge 1.53) + P(z \le -1.53) = p = 0.126 > 0.05$，此时 $p > \alpha$，因此我们无法拒绝零假设。

## 区间估计与假设检验的关系

之前我们学过如何计算总体均值的置信区间，在 $\sigma$ 已知的情况下，总体均值 $100 \times (1-\alpha)\%$ 的置信区间估计为：

$$\displaystyle \bar{x}\pm z_{\alpha/2} \frac{\sigma}{\sqrt n}$$

此时原假设 $H_0: \mu = \mu_0$，备择假设 $H_{\alpha}: \mu \neq \mu_0$

从总体中抽取一个简单随机样本 并利用样本均值 $\bar x$ 建立总体均值 $\mu$ 的置信区间

如果置信区间包含假设值 $\mu_0$，则不能拒绝$H_0$，否则拒绝$H_{\alpha}$

已知
- $\mu_0 = 295$
- $\sigma = 12$
- $n = 50$
- $\bar{x} = 297.6$
- 显著性水平 $\alpha = 0.05$，对应的 $z_{\alpha/2} = 1.96$

此时置信区间的估计为：

$$\displaystyle 297.6 \pm 1.96 \times \frac{12}{\sqrt{50}}$$

$$\displaystyle 297.6 \pm 1.96 \times 3.3$$

区间为：$$[294.3, 300.9]$$

根据这个结果，我们可以有 95% 的把握确信，总体均值 $\mu$ 位于区间 $[294.3, 300.9]$ 之间，

由于 $\mu_0 = 295$ 位于区间之内，所以我们不能拒绝零假设。

```python
alpha = 0.05

mu_0 = 295
x_bar = 297.6
sigma = 12
n = 50

stdev = sigma/n**0.5

# Our calculated z-score in this example:
p_crit = norm.ppf(q = 1 - alpha/2)
p_val = norm.cdf(x = x_bar, loc = mu_0, scale = stdev)
p_stat = norm.ppf(p_val)
#print(p_crit, p_val, p_stat)
```

```python
def two_tail_test(p_crit, p_stat, p_val):
    x = np.linspace(start = norm.ppf(0.001), stop = norm.ppf(0.999), num = int(1e5))
    hx = norm.pdf(x)
    # Plot the density
    fig, ax = plt.subplots(num = 2, figsize = (10, 8))
    _ = plt.plot(x, hx, color = "black")
    _ = plt.margins(y = 0)
    _ = plt.ylim((0, 0.41))

    # Shade the probability p-value
#    _ = plt.fill_between(x[x <= p_stat], hx[x <= p_stat], edgecolor = None, color = clrs.to_rgba(c = (0, 1, 0, 0.4)))
    # plot the observed t-statictic:
    _ = plt.axvline(x = p_stat, color = "darkgreen", linestyle = "--", linewidth = 10)

    # Shade the probability alpha / 2
    _ = plt.fill_between(x[x <= -np.abs(p_crit)], hx[x <= -np.abs(p_crit)], 
                     edgecolor = "red", linestyle = "-", linewidth = 2,
                     facecolor = clrs.to_rgba(c = (1, 0, 0, 0.1)), zorder = 10)

    # Shade the probability 1 - alpha / 2
    _ = plt.fill_between(x[x >= np.abs(p_crit)], hx[x >= np.abs(p_crit)], 
                     edgecolor = "red", linestyle = "-", linewidth = 2,
                     facecolor = clrs.to_rgba(c = (1, 0, 0, 0.1)), zorder = 10)
    
    # plot the critical value -|t_c| and t_c|:
    _ = plt.axvline(x = -np.abs(p_crit), color = "red", linestyle = "--")
    _ = plt.axvline(x = np.abs(p_crit), color = "red", linestyle = "--")
    #
    _ = ax.set_xticks([-np.abs(p_crit), p_stat, np.abs(p_crit)])
    _ = ax.set_xticklabels(["$-|t_c|$", "$t_{stat}$", "$|t_c|$"])
    ax.get_xticklabels()[0].set_color("red")
    ax.get_xticklabels()[1].set_color("darkgreen")
    ax.get_xticklabels()[2].set_color("red")
    _ = plt.title("p-value for the two-tail test; $H_1: \\beta_1 \\neq 0$")
    plt.show()
```

```python
two_tail_test(p_crit, p_stat, p_val)
```

# 总体均值的检验：$\sigma$ 未知情形

如果总体标准差 $\sigma$ 未知，因此我们不能使用正态分布进行假设检验。

先回顾一下总体均值假设检验的检验统计量：$\sigma$ 已知的情形，计算 $z$ 值的公式为：

$$z = \frac {\bar{x}- \mu_0} {\frac{\sigma}{\sqrt{n}}}$$

- $\bar{x}$ 为样本均值
- $\mu_0$ 为原假设均值
- $\sigma$ 为总体标准差
- $n$ 样本数
- $z$ 为检验统计量

如果总体标准差 $\sigma$ 未知，检验统计量（$z$ 分数）不服从正态分布，但是还是服从一个新的分布 $t$ 分布，此时检验统计量的计算方式如下：

$$t = \frac {\bar{x}- \mu_0} {\frac{s}{\sqrt{n}}}$$

- $\bar{x}$ 为样本均值
- $\mu_0$ 为原假设均值
- $s$ 为样本标准差
- $n$ 样本数
- $t$ 为检验统计量

由于计算方法与总体均值已知完全一样，只是原来需要查 $z$ 值表，现在需要查 $t$ 值表，所以这里就不赘述了。

```python
alpha = 0.05

mu_0 = 295
x_bar = 297.6
sigma = 12
n = 50

stdev = sigma/n**0.5

# Our calculated z-score in this example:
t_crit = t.ppf(q = 1 - alpha/2, df = n-2)

t_val = t.cdf(x = x_bar, loc = mu_0, scale = stdev, df = n-2)
t_stat = t.ppf(t_val, df = n-2)
print(t_crit, t_val, t_stat)
```

```python
# 双侧 t 检验
x = np.linspace(start = t.ppf(0.001, df = n-2), stop = t.ppf(0.999, df = n-2), num = int(1e5))
hx = t.pdf(x, df = n-2)
# Plot the density
fig, ax = plt.subplots(num = 2, figsize = (10, 8))
_ = plt.plot(x, hx, color = "black")
_ = plt.margins(y = 0)
_ = plt.ylim((0, 0.41))
# Shade the probability p-value
_ = plt.fill_between(x[x <= t_stat], hx[x <= t_stat], edgecolor = None,
                 color = clrs.to_rgba(c = (0, 1, 0, 0.4)))
# plot the observed t-statictic:
_ = plt.axvline(x = t_stat, color = "darkgreen", linestyle = "--")
# Shade the probability alpha / 2
_ = plt.fill_between(x[x <= -np.abs(t_crit)], hx[x <= -np.abs(t_crit)], 
                 edgecolor = "red", linestyle = "-", linewidth = 2,
                 facecolor = clrs.to_rgba(c = (1, 0, 0, 0.1)), zorder = 10)
# Shade the probability 1 - alpha / 2
_ = plt.fill_between(x[x >= np.abs(t_crit)], hx[x >= np.abs(t_crit)], 
                 edgecolor = "red", linestyle = "-", linewidth = 2,
                 facecolor = clrs.to_rgba(c = (1, 0, 0, 0.4)), zorder = 10)
# plot the critical value -|t_c| and t_c|:
_ = plt.axvline(x = -np.abs(t_crit), color = "red", linestyle = "--")
_ = plt.axvline(x = np.abs(t_crit), color = "red", linestyle = "--")
#
_ = ax.set_xticks([-np.abs(t_crit), t_stat, np.abs(t_crit)])
_ = ax.set_xticklabels(["$-|t_c|$", "$t_{stat}$", "$|t_c|$"])
ax.get_xticklabels()[0].set_color("red")
ax.get_xticklabels()[1].set_color("darkgreen")
ax.get_xticklabels()[2].set_color("red")
_ = plt.title("p-value for the two-tail test; $H_1: \\beta_1 \\neq 0$")
plt.show()
```

# 总体比率

之前我们假设检验的变量都是离散型随机变量，现在我们也可以推广到连续型随机变量。

回顾一下，$E(\bar p_0) = p$ 时，无限总体 $\bar p$ 的标准差为：

$$\sigma_{\bar p} = \sqrt{\frac {p_0(1-p_0)}{n}}$$

总体比率假设检验的检验统计量：

$$z = \frac{\bar p - p_0}{\sigma_{\bar p}}$$

假设总体女性高尔夫球员的比率 $p_0 = 0.2$，我们现在随机选取 400 名高尔夫球员组成一个随机样本，其中 100 名球员是女性，样本中女性高尔夫球员的比率为：

$$\bar{p} = \frac{100}{400} = 0.25$$

样本标准差为：

$$\sigma_{\bar p} = \sqrt{\frac {p_0(1-p_0)}{n}} = \sqrt{\frac {0.2(1-0.2)}{400}} = 0.02$$

总体比率假设检验的检验统计量：

$$z = \frac{\bar p - p_0}{\sigma_{\bar p}} = \frac {0.25 - 0.2}{0.02} = 2.5$$

查询 $z = 2.5$ 时的 $p = 0.9938$，$P(z \le 2.5) = 1 - 0.9938 = 0.0062$

设定显著性水平 $\alpha = 0.05$，此时 $P(z \le 2.5) < \alpha$，因此我们可以拒绝零假设。

# 计算第二类错误的概率
先让我们复习一下第二类错误：$H_{\alpha}$ 为真时，拒绝 $H_{\alpha}$

下面我们用一个货物接收的抽样问题，来解释如何计算第二类错误。例如，对供应商的一批电池，某名质量控制管理人员必须决定是接受这批货物，还是因其质量差而将货物退还供应商。

假定设计规格中要求供应商电池的平均使用寿命至少为 120 个小时。为了评估这批货物的质量，我们选取 36 节电池组成样本进行检验，根据样本来决定是接受这批货物，还是因其质量差而将这批货物退还供应商。

令 $\mu$ 代表这批电池使用寿命的均值（单位：小时），对总体均值建立如下形式的原假设和备择假设。
$$H_0: \mu \ge 120\\
H_\alpha: \mu < 120$$

如果拒绝 $H_0$，则得出备择假设为真的结论。这一结论表明，将这批货物退还给供应商是明智之举。但是，如果不能拒绝 $H_0$，决策者仍需确定采取某种措施。

所以，当没有直接得出“ $H_0$ 为真”的结论，而仅仅是“不能拒绝 $H_0$ ”时，决策者将认为这批货物的质量符合标准而决定接收这批货物。

在这种决策中，我们建议将假设检验过程扩展，对发生第二类错误的概率予以控制。由于当不能拒绝 $H_0$ 时，必须做出决策并采取措施，所以有关发生第二类错误的概率的信息对我们很有帮助。

总结一下已知条件：

已知
- $\mu_0 = 120$
- $\sigma = 12$
- $n = 36$
- 显著性水平 $\alpha = 0.05$，对应的 $z_{\alpha} = 1.645$


$$\sigma_\bar{x} = \frac{\sigma}{\sqrt{n}} = \frac{12}{\sqrt{36}} = 2$$

$$z = \frac {\bar{x}- \mu_0} {\sigma_\bar{x}} = \frac {\bar{x}- 120}{2} \le -1.645$$

求解得，当

$$\bar{x} \le 120-1.645 \times 2 = 116.71$$

时，拒绝 $H_0$，只要我们抽样的均值 $\bar{x} > 116.71$ 时，就应该接受货物。

但是，实际情况有可能是，总体的均值有可能小于 120 小时，但是我们却因为计算结果，接受了这批货物，这种情况我们就犯了第二类错误。

要想计算第二类错误的概率，我们需要选择一个小于 120 的值，假定其为总体均值。

假定真正的总体均值 $\mu = 112$，则第二类错误就是：**当实际的总体均值 $\mu = 112$ 时，样本均值 $\bar{x} > 116.71$ 的概率。**

$$z = \frac {\bar{x}- \mu} {\sigma_\bar{x}} = \frac {116.71 - 112}{2} = 2.36$$

此时 $z$ 分数对应的 $p$ 值为 0.9909，

我们把当 $H_0$ 为假时，得出拒绝 $H_0$ 的正确结论的概率为检验的**功效**，这里的功效为 0.9909

令 $\beta$ 表示发生第二类错误的概率，则：

$$\beta = 1 - p = 1 - 0.9909 = 0.0091$$

当实际的总体均值 $\mu = 112$ 时，样本均值 $\bar{x} > 116.71$ 的概率非常小，仅为 0.91%。

总之，对总体均值进行假设检验时，可以按照如下所给定的流程逐步计算发生第二类错误的概率：

1. 确立原假设和备择假设。
2. 在显著性水平 $\alpha$ 下，根据临界值方法确定临界值并建立检验的拒绝规则。
3. 利用（2）中所得的拒绝规则，求解与检验统计量的临界值相对应的样本均值的取值。
4. 利用步骤（3）中的结果，得到接受 $H_0$ 时所对应的样本均值的值，这些值构成了检验的接受域。
5. 对于满足备择假设的 $\mu$ 值，利用 $\bar{x}$ 的抽样分布和步骤（4）中的接受域，计算样本均值落在接受域的概率。这一概率值即为在选定的 $\mu$ 值处发生第二类错误的概率。

# 总体均值单侧假设检验的样本容量

$$n = \frac {(z_{\alpha} + z_{\beta})^2 \sigma^2} {(u_0-u_{\alpha})^2}$$

还是以上面的电池检验题目为例，设计规格要求电池使用寿命的均值至少为 120 个小时。

如果 $H_0:\mu≥120$ 被拒绝，则该批货物被拒收。

我们假定质量控制管理人员要求第一类错误和第二类错误的可接受概率达到如下要求。
- 关于第一类错误的要求：如果货物中电池寿命的均值为 $\mu=120$，那么甘愿冒 $\alpha=0.05$ 的风险概率拒绝这批货物。
- 关于第二类错误的要求：如果货物中电池寿命的均值比规格要求少 5 小时（即 $\mu=115$），那么甘愿冒 $\beta=0.10$ 的风险概率接受这批货物。

如上要求是基于管理者的判断得到的。不同的人可能会对此概率做出不同的限制。但是，在确定样本容量之前，必须明确允许发生两类错误概率的大小。

在例子中，$\alpha=0.05$，$\beta=0.10$。利用标准正态概率分布，有 $z_{0.05}=1.645$， $z_{0.1}=1.28$。

从关于错误概率的要求中，我们已知我们期望的总体均值 $\mu_0=120$，实际的总体均值 $\mu_a=115$。

最后，假定已知总体标准差 $\sigma=12$。在货物接收的例子中，我们建议使用的样本容量为：

$$n = \frac{(z_{\alpha} + z_{\beta})^2\sigma^2}{(u_0-u_{\alpha})^2}
 = \frac{(1.645 + 1.28)^2 \times 12^2}{(120-115)^2} = 49.3$$

建议的样本容量为 50 个。
